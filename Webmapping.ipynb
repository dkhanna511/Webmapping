# Mapbox Style specification and Adding markers

A Mapbox style is a document that defines the visual appearance of a map: what data to draw, the order to draw it in, and how to style the data when drawing it. A style document is a JSON object with specific root level and nested properties. This specification defines and describes these properties.

```python
#for changing the background of the map
map=fl.Map(location=[28.567600,77.241298], tiles="Mapbox Bright")   #location:lattitude,longitude
map.save("map1.html")
help(fl.Marker)
#for creating a method (add_child)
map.add_child(fl.Marker(location=[28.58,77,98], popup="hi I am a marker", icon=fl.Icon(color="green")))
map.save("Map1.html")
```

### Creating Feature Groups



```python

#for changing the background of the map to bright color
map=fl.Map(location=[28.567600,77.241298], tiles="Mapbox Bright")   #locatuion:lattitude,longitude

#for adding markers the the map, method to add children

#for creating a feature group to the map
fg=fl.FeatureGroup("My Map")
fg.add_child(fl.Marker(location=[28.58,77,98], popup="hi I am a marker", icon=fl.Icon(color="green")))
map.add_child(fg)
map.save("Map1.html")
```

```python
#adding multiple markers in a map
fg=fl.FeatureGroup(name="My map")

for coordinates in [[28.55,77.1],[27.652,78.21]]:
    fg.add_child(fl.Marker(location=coordinates, popup="Hello", icon=fl.Icon(color='green')))
    
map.add_child(fg)
map.save("Map1.html")
```

# Adding markers with data files

# Creating Popup windows for map features

till above code we have the same popup for every map feature.
Here we will create different popup windows for different locations over the map in order to conver some information
We will be adding the elevation data for every location here. 

```python

data=pd.read_csv("Volcanoes_USA.txt")
#print(data)

#converting the data framweworks in the form of list as working on list is faster
lat=list(data["LAT"])
lon=list(data["LON"])
elev=list(data["ELEV"])
fg=fl.FeatureGroup(name="My map")


#here, str function is used since type(elev) is of float64 type which is not accepted as an argument there. 
for lt, ln, el in zip(lat, lon, elev):
    fg.add_child(fl.Marker(location=[lt, ln], popup=str(el)+"m", icon=fl.Icon(color='green')))
    
map.add_child(fg)
map.save("Map1.html")
```

# Color based point markers

```python
data=pd.read_csv("Volcanoes_USA.txt")
#print(data)

#converting the data framweworks in the form of list as working on list is faster
lat=list(data["LAT"])
lon=list(data["LON"])
elev=list(data["ELEV"])
fg=fl.FeatureGroup(name="My map")


#here, str function is used since type(elev) is of float64 type which is not accepted as an argument there. 
for lt, ln, el in zip(lat, lon, elev):
    if(el<1000):
        clr='green'
    elif(el>1000 and el<2000):
        clr='orange'
    else:
        clr='red'
    fg.add_child(fl.Marker(location=[lt, ln], popup=str(el)+"m", icon=fl.Icon(color=clr)))
    
map.add_child(fg)
map.save("Map1.html")

#dir(fl)
#help(fl.Circle)
```

# Adding and styling circle markers

```python
data=pd.read_csv("Volcanoes_USA.txt")
#print(data)

#converting the data framweworks in the form of list as working on list is faster
lat=list(data["LAT"])
lon=list(data["LON"])
elev=list(data["ELEV"])
fg=fl.FeatureGroup(name="My map")


#here, str function is used since type(elev) is of float64 type which is not accepted as an argument there. 
for lt, ln, el in zip(lat, lon, elev):
    if(el<1000):
        clr='green'
    elif(el>1000 and el<2000):
        clr='orange'
    else:
        clr='red'
    fg.add_child(fl.CircleMarker(location=[lt, ln],radius=6, popup=str(el)+"m",fill_color=clr,color='grey', fill_opacity=0.7))
    
map.add_child(fg)
map.save("Map1.html")

#dir(fl)
#help(fl.Circle)
```

# Colour Based Polygon Features

```python
data=pd.read_csv("Volcanoes_USA.txt")
#print(data)

#converting the data framweworks in the form of list as working on list is faster
lat=list(data["LAT"])
lon=list(data["LON"])
elev=list(data["ELEV"])
fg=fl.FeatureGroup(name="My map")


#here, str function is used since type(elev) is of float64 type which is not accepted as an argument there. 
for lt, ln, el in zip(lat, lon, elev):
    if(el<1000):
        clr='green'
    elif(el>1000 and el<2000):
        clr='orange'
    else:
        clr='red'
    fg.add_child(fl.CircleMarker(location=[lt, ln],radius=6, popup=str(el)+"m",fill_color=clr,color='grey', fill_opacity=0.7))

fg.add_child(fl.GeoJson(data=open('World.json', 'r', encoding='utf-8-sig').read(), 
style_function=lambda x:{'fillColor':'yellow' if x['properties']['POP2005']<=1000000
else 'orange' if 1000000<x['properties']['POP2005']<=2000000
else 'green'}))    
map.add_child(fg)
map.save("Map1.html")

#dir(fl)
#help(fl.Circle)
```

# Adding a layer control panel

In layer control panel , we make seperate feature groups for population data set and volcanoes data set. 
After that we apply layer control method by which we can shift to which layer(feature group) we want at the time of execution of the map in sumultaneous process. 

```python
data=pd.read_csv("Volcanoes_USA.txt")
#print(data)

#converting the data framweworks in the form of list as working on list is faster
lat=list(data["LAT"])
lon=list(data["LON"])
elev=list(data["ELEV"])

#feature group for volvanoes denoted by fgv
fgv=fl.FeatureGroup(name="Volcanoes")


#here, str function is used since type(elev) is of float64 type which is not accepted as an argument there. 
for lt, ln, el in zip(lat, lon, elev):
    if(el<1000):
        clr='green'
    elif(el>1000 and el<2000):
        clr='orange'
    else:
        clr='red'
    fgv.add_child(fl.CircleMarker(location=[lt, ln],radius=6, popup=str(el)+"m",fill_color=clr,color='grey', fill_opacity=0.7))

#feature group for population denoted by fgp
fgp=fl.FeatureGroup(name="Population")
fgp.add_child(fl.GeoJson(data=open('World.json', 'r', encoding='utf-8-sig').read(), 
style_function=lambda x:{'fillColor':'yellow' if x['properties']['POP2005']<=1000000
else 'orange' if 1000000<x['properties']['POP2005']<=2000000
else 'green'})) 


map.add_child(fgv)
map.add_child(fgp)
map.add_child(fl.LayerControl())
map.save("Map1.html")

#dir(fl)
#help(fl.Circle)
```
